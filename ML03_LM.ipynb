```python
#線形モデル(linear model)

#ライブラリのインストール p11
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
import mglearn
from sklearn.model_selection import train_test_split

#In[24]--------------------------
#１次元の waveデータセットで w[0]とbを求める
mglearn.plots.plot_linear_regression_wave() # regressionは回帰
```

```python
# 線形回帰(linear regression)(通常最小二乗法(ordinary least squares:OLS))
#In[25]----------------------------
from sklearn.linear_model import LinearRegression
# 引数n_samplesで指定した個数の点について1つの特徴量とターゲットの値を持ち、回帰を扱うのに適している。
X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train) # 訓練データと訓練ターゲットを用いてモデルを学習(fitで学習)させる

#In[26]----------------------------
print("lr.coef_:{}".format(lr.coef_)) # パラメータ w(重み、係数)を表示, coef_属性は入力特徴量ごとに1つの要素を持つNumpy配列となる
print("lr.intercept_:{}".format(lr.intercept_)) # 切片
#訓練データから得られた属性には_をつける
```

```python
#In[27]----------------------------
print("Training set score:{:.2f}".format(lr.score(X_train, y_train)))
print("Test set score:{:.2f}".format(lr.score(X_test, y_test)))
```

```python
#In[28]----------------------------
X, y = mglearn.datasets.load_extended_boston()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)

#In[29]----------------------------
print("Training set score:{:.2f}".format(lr.score(X_train, y_train)))
print("Test set score:{:.2f}".format(lr.score(X_test, y_test)))
```

```python
#リッジ回帰(ridge regression)
#In[30]----------------------------
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print("Training set score:{:.2f}".format(ridge.score(X_train, y_train)))
print("Test set score:{:.2f}".format(ridge.score(X_test, y_test)))
```

```python
#In[31]---------------------------
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Training set score:{:.2f}".format(ridge10.score(X_train, y_train)))
print("Test set score:{:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
#In[32]--------------------------
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Training set score:{:.2f}".format(ridge01.score(X_train, y_train)))
print("Test set score:{:.2f}".format(ridge01.score(X_test, y_test)))
```

```python
#In[33]--------------------------
plt.plot(ridge.coef_, 's', label="Ridge alpha=1") # Ridge alpha=1のデータポイントの形を指定
plt.plot(ridge10.coef_, '^', label="Ridge alpha=10")  # Ridge alpha=10のデータポイントの形を指定
plt.plot(ridge01.coef_, 'v',label="Ridge alpha=0.1") # Ridge alpha=0.1のデータポイントの形を指定

plt.plot(lr.coef_, 'o', label="LinearRegression") # LinarRegressionのデータポイントの形を指定
plt.xlabel("Coefficient index")　#特徴量インデックス
plt.ylabel("Coefficient magnitude") #係数の大きさ
plt.hlines(0, 0, len(lr.coef_))
plt.ylim(-25, 25)
plt.legend()
```
